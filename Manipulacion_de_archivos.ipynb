```julia
open("escribir_stream.csv", "w") do f2
    for line in 1:5
        write(f2, readline(f))
    end
end
```

```julia
;cat escribir_stream.csv
```

```julia
;wc -l escribir_stream.csv
```

```julia
open("escribir_stream.csv", "a") do f2
    for line in 1:5
        write(f2, readline(f))
    end
end
```

```julia
;cat escribir_stream.csv
```

```julia
open("escribir_stream.csv", "a") do f2
    for line in 1:5
        write(f2, readline(f)*"\n")
    end
end
```

```julia
close(f2)
```
